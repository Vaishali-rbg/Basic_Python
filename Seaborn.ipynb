```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

sns.set(style="white")
rs = np.random.RandomState(10)
d = rs.normal(size=100)
sns.distplot(d, kde=True, color="m")
plt.show()
```

```python
sns.set(style="dark")
fmri = sns.load_dataset("fmri")
sns.lineplot(x="timepoint",
             y="signal",
             hue="region",
             style="event",
             data=fmri)
plt.show()
```

```python
#
sns.set(style="ticks")
df = sns.load_dataset("anscombe")
sns.lmplot(x="x", y="y", data=df)
plt.show()
```

```python
#stripplot
x =['sun', 'mon', 'fri', 'sat', 'tue', 'wed', 'thu']
y =[5, 6.7, 4, 6, 2, 4.9, 1.8]
 

ax = sns.stripplot(x);
ax.set(xlabel ='Days', ylabel ='Amount_spend')
plt.title('Strip Plot');
plt.show()


ax = sns.stripplot(y);
ax.set(xlabel ='Days', ylabel ='Amount_spend')
plt.title('Strip Plot');
plt.show()
```

```python
df = sns.load_dataset('tips')
sns.stripplot(x='day',y = 'total_bill', hue = 'smoker', data = df,dodge=True)
```

```python
#swarmplot
#first style
sns.set(style="whitegrid")
iris = sns.load_dataset('iris')
 
ax = sns.swarmplot(x='species', y='sepal_length', data=iris)
plt.title('Swarm Plot')    
plt.show()
```

```python
#second style
sns.set(style="darkgrid")
 
iris = sns.load_dataset('iris')
ax = sns.swarmplot(x='sepal_length', y='species', data=iris)
plt.title('Swarm Plot')    
plt.show()
```

```python
#Barplot
import seaborn as sns
 

df = sns.load_dataset('tips')
sns.set(style="darkgrid") 
sns.barplot(x ='sex', y ='total_bill', data = df, 
            palette ='plasma')
```

```python
#Countplot
import seaborn as sns
 
df = sns.load_dataset('tips')
sns.set(style="darkgrid")
sns.countplot(x ='sex', data = df)
```

```python
#Boxplot
import seaborn as sns
df = sns.load_dataset('tips')
sns.set(style="darkgrid")
sns.boxplot(x='day', y='total_bill', data=df, hue='smoker')
```

```python
#Violinplot
import seaborn as sns

df = sns.load_dataset('tips')
sns.set(style="ticks")
sns.violinplot(x='day', y='total_bill', data=df,
               hue='sex', split=True)
```

# MULTIPLOTTING

```python
#FacetGrid
import seaborn
import matplotlib.pyplot as plt

df = seaborn.load_dataset('tips')
graph = seaborn.FacetGrid(df, col ="sex",  hue ="day")
graph.map(plt.scatter, "total_bill", "tip", edgecolor ="w").add_legend()
plt.show()
```

```python
#PairGrid
import seaborn
import matplotlib.pyplot as plt

df = seaborn.load_dataset('tips')
graph = seaborn.PairGrid(df, hue ='day')
graph = graph.map_diag(plt.hist)
graph = graph.map_offdiag(plt.scatter)
graph = graph.add_legend()
plt.show()
```

```python
#RegressionPlot

import seaborn
import matplotlib.pyplot as plt

df = seaborn.load_dataset('tips')
graph = seaborn.FacetGrid(df, col ='time', hue ='smoker')
graph.map(seaborn.regplot, "total_bill", "tip").add_legend()
plt.show()
```

```python
#Factorplot(catplot)
import seaborn as sns


df = sns.load_dataset('tips')
sns.catplot(x ='size', y ='tip', 
               hue = 'sex', data = df)
plt.show()
```
