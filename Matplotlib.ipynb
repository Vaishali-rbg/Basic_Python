```python
x = [3, 1, 3] 
y = [3, 2, 1] 
z = [1, 3, 1] 
plt.figure() 
   
# addind first subplot 
plt.subplot(121)  # Single argument to subplot must be a three-digit integer
plt.plot(x, y) 
   
# addding second subplot 
plt.subplot(122) 
plt.plot(z, y)
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].plot([1, 2, 3, 4], [1, 2, 3, 4])

# plotting the data in the 1st subplot only
axes[0].plot([1, 2, 3, 4], [4, 3, 2, 1])

# plotting the data in the 2nd subplot only
axes[1].plot([1, 2, 3, 4], [1, 1, 1, 1])
```

```python
x = [3, 1, 3] 
y = [3, 2, 1] 
z = [1, 3, 1] 
   
# adding the subplots
axes1 = plt.subplot2grid ((6, 1), (0, 0), rowspan = 2,  colspan = 2) 
axes2 = plt.subplot2grid ((6, 1), (2, 0), rowspan = 2, colspan = 2) 
axes3 = plt.subplot2grid ((6, 1), (4, 0), rowspan = 2, colspan = 2) 
   
# plotting the data
axes1.plot(x, y)
axes2.plot(x, z)
axes3.plot(z, y)
```

Types of plots:
Line Graph,
Bar Chart,
Histograms,
Scatter Plot,
Pie-Chart,
3D Plots.

```python
import matplotlib.pyplot as plt

x = [3, 1, 3] 
y = [3, 2, 1] 

plt.plot(x, y)
plt.title("Line Chart")

plt.legend(["Line"])
plt.show()
```

```python
x = [3, 1, 3, 12, 2, 4, 4] 
y = [3, 2, 1, 4, 5, 6, 7] 

plt.bar(x, y)
plt.title("Bar Chart")

plt.legend(["bar"])
plt.show()
   
```

```python
 x= [1, 2, 3, 4, 3, 6, 5, 4] 
 
plt.hist(x, bins = [1, 2, 3, 5, 5, 6, 7])

plt.title("Histogram")

plt.legend(["bar"])
plt.show()
```

```python
x = [3, 1, 3, 7, 2, 4, 4]
y = [2, 4, 5, 7, 8, 11, 3]
   
# This will plot a simple scatter chart
plt.scatter(x, y)
   
# Adding legend to the plot
plt.legend(["scatter"])
   
# Title to the plot
plt.title("Scatter chart")
plt.show()
```

```python
x = [1, 2, 3, 4] 

e  =(0.1, 0, 0, 0)

plt.pie(x, explode = e)

plt.title("Pie chart")
plt.show()
```
